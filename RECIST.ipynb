Response Evaluation Criteria in Solid Tumors, 종양 반응을 평가하기 위한 가이드라인 <br><br>

|RECIST 가이드라인|RECIST 1.0|RECIST 1.1|
|------|---|---|
|대상 병변 수|기관당 최대 5개, 총 10개까지|기관당 최대 2개, 총 5개까지|
|LN 평가|명확한 지침 없음|(1) 단축 >=15mm인 경우 표적 병변 <br>(2) 10mm <= 단축 <= 15mm인 경우 비표적 병변<br> (3) 단축 < 10mm인 경우 비 병리학|
|질병 진행의 설명|대상병변 합계의 20% 증가(최소 절대 크기 증가 없음)|대상 병변의 합계 20% 증가와 5mm 절대 증가 필요|


### Paper

**Validity of RECIST Version 1.1 for Response Assessment in Metastatic Cancer: A Prospective, Multireader Study** <br><br>



*   전이성 암 환자에게 RECIST 1.1을 사용하여 표적 병변 선택과 치료 반응 사이의 관계를 조사한 연구<br>
*  세명의 방사선 전문의가 각각 해석하여 판독자 간의 일치도 분석<br>
* 316명의 데이터 중 128명에서만 동일한 표적 병변을 선택하고, 188명에서는 다르게 선택함<br>
* 판독자 각각이 선택한 표적 병변이 일치할때는 일치도가 높은 편이지만, 그렇지 않은 경우에는 일치도가 감소함.<br><br>


결론:: RECIST 1.1의 표적 병변이 전체 종양의 부하나 치료 반응을 충분히 반영하지 못할 수 있음.  

**Twenty Years On: RECIST as a Biomarker of Response in Solid Tumors an EORTC Imaging Group -- ESOI Joint Paper<br><br>**
* RECIST는 완벽하지 않을 수 있지만, 항암제에 대한 과학적인 근거를 제공하는데 여전히 유효함
* 재현성은 측정의 변동성에 영향을 받으며, 판독 경험과 대상 병변 선택, 새로운 병변의 감지 및 해석에 영향을 받을 수 있음
* 일부 상황에서 치료 유발 변화를 정확히 평가하지 못할 수도 있음

결론:: 측정 자동화는 결과의 변동성을 감소시킬 수 있으나, 기술적 어려움과 사람의 판단이 필요할 수 있음
