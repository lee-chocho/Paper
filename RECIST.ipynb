### RECIST

**RECIST란?**

Response Evaluation Criteria in Solid Tumors, 종양 반응을 평가하기 위한 가이드라인 <br><br>

|RECIST 가이드라인|RECIST 1.0|RECIST 1.1|
|------|---|---|
|Target Lesion 수|기관당 최대 5개, 총 10개까지|기관당 최대 2개, 총 5개까지|
|Lesion 기준|장축 기준 10mm 이상|장축 기준 10mm 이상|
|림프 노드 평가|명확한 가이드라인이 제공되지 않음|단축 측정 사용 및 기록|
|PD(progressive disease)|대상병변 합계의 20% 증가(최소 절대 크기 증가 없음)|대상 병변의 합계 20% 증가와 5mm 절대 증가 필요|
|PR(partial response)|대상 병변 장축의 합계 30% 감소|대상 병변 장축의 합계 30% 감소|
|CR(complete response)|암의 증거가 모두 사라짐|암의 증거가 모두 사라짐|
|SD(stable disease)|위의 어느것도 해당되지 않음|위의 어느것도 해당되지 않음|

### Paper

**Validity of RECIST Version 1.1 for Response Assessment in Metastatic Cancer: A Prospective, Multireader Study** <br><br>



*   전이성 암 환자에게 RECIST 1.1을 사용하여 표적 병변 선택과 치료 반응 사이의 관계를 조사한 연구<br>
*  세명의 방사선 전문의가 각각 해석하여 판독자 간의 일치도 분석<br>
* 316명의 데이터 중 128명에서만 동일한 표적 병변을 선택하고, 188명에서는 다르게 선택함<br>
* 판독자 각각이 선택한 표적 병변이 일치할때는 일치도가 높은 편이지만, 그렇지 않은 경우에는 일치도가 감소함.<br><br>


결론:: RECIST 1.1의 표적 병변이 전체 종양의 부하나 치료 반응을 충분히 반영하지 못할 수 있음.  

<br><br>**Twenty Years On: RECIST as a Biomarker of Response in Solid Tumors an EORTC Imaging Group -- ESOI Joint Paper<br><br>**
* RECIST는 완벽하지 않을 수 있지만, 항암제에 대한 과학적인 근거를 제공하는데 여전히 유효함
* 재현성은 측정의 변동성에 영향을 받으며, 판독 경험과 대상 병변 선택, 새로운 병변의 감지 및 해석에 영향을 받을 수 있음
* 일부 상황에서 치료 유발 변화를 정확히 평가하지 못할 수도 있음

결론:: 측정 자동화는 결과의 변동성을 감소시킬 수 있으나, 기술적 어려움과 사람의 판단이 필요할 수 있음

<br><br>**RECIST Needs Revision: A Wake-up Call for Radiologists**<br><br>

* Coy 박사 등은 RECIST 컨소시엄이 비정상적인 이유(즉, 비표적 종양 진행 또는 새로운 병변)으로 인해 진행성 질환으로 분류된 환자의 생존 기간(progression-free survival)이 target lesions의 진행으로 인해 진행성 질환으로 분류된 환자보다 짧은 것을 발견
* 이는 RECIST 컨소시엄이 표적 병변 외부의 진행성 질병을 드물게 발생한다고 가정한 것과 다른 결과로, Coy등이 분석한 결과는 아래와 같다.
        (1) 데이터웨어하우스 분석은 선택된 대상 병변의 결과만을 시뮬레이션하므로 다른 표적 병변의 결과를 예측할 수 없음. 이것은 결과적으로 표적 병변의 선택이 반응 분류에 영향을 미칠 수 있다는 것을 의미함.
        (2) RECIST 컨소시엄은 대상 병변을 선택하는 방법에 대한 명확한 기준을 제시하지 않고 보고 가능한 표적 병변의 최대 수를 제한함. 이로 인해 대부분의 RECIST 분석은 1~2개의 표적 병변에서만 수행되고 다른 병변은 고려되지 않을 수 있음.

* 즉, RECIST 시스템이 대상 병변 외부의 진행성 질병을 고려하지 않고 표적 병변의 선택에 의존하여 결과를 분석하기 때문에 비정상적인 결과가 나올 수 있다고 결론지음.

결론:: 방사선과 전문의가 표적 병변 이외의 진행성 질병을 찾기 위해 모든 병변을 고려해야 할 것과 최신 반응 평가 소프트웨어 도구의 사용으로 RECIST 반응 분석의 정확성과 재현성을 향상시킬 수 있음.


<br><br>**Evaluation of the capability and reproducibility of RECIST 1.1. measurements by technologists in breast cancer follow-up: a pilot study**<br><br>

* 본 연구에서 표적 병변이 기준선 CT에서 방사선과 전문의(radiologists)에 의해 식별된 유방암 환자의 후속 조치에서 치료 후 CT에서 technologists가 수행한 RECIST 1.1 측정의 분류를 평가함.
* DB 수집 기간 동안 radiologists가 제시한 RECIST 1.1 반응 분류는 CR 5건, PR 60건, SD 93건, PD 112건 (총 270건)
* technologists에 의해 측정된 mm 및 백분율 측정의 차이는 중앙값이 0에 가깝고 분산이 낮아 전체적으로 좋은 결과를 제공했지만, 극단값을 보임
* 극단값의 경우는 CT 재구성 선택의 오류로 설명되며, 결과적으로 슬라이스 두께와 병변 크기의 차이로 인해 발생 <br>

결론:: 일부 숙련된 technologists가 질병 진행을 잘 식별하고 재현성이 뛰어난 RECIST 1.1 기준에 따라 측정을 수행할 수 있는 것으로 분석함.
