## The relationship between COPD and lung cancer

[논문링크](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4718929/)<br>

* COPD와 폐암은 모두 흡연으로 인해 발생하는 주요 건강 문제이며, 전 세계적으로 질병 부담이 크다.
* COPD와 폐암은 흡연에 의해 독립적으로 유발되는 경우보다 동반 질환으로써 더 잘 발생한다.
* COPD는 폐암, 특히 편평 세포 암종에 대한 독립적인 위험 인자이다.
* 대부분의 흡연자에서 흡연 부담은 복구 메커니즘에 의해 잘 처리되지만, 만약 실패하여 돌연변이가 발생하면 암으로 이어지고, 세포와 단백질에 대한 손상이 너무 커지면 COPD가 발생한다.
* COPD는 산화 스트레스 증가와 그게 따른 DNA 손상, 전염증성 사이토카인에 대한 만성 노출, DNA 복구 메커니즘의 억제 및 세포 증식 증가를 통해 폐암의 원동력이 될 수 있다.
* 폐암 발생률은 흡연자의 경우 남성 17.2%, 여성 11.6%이며 비흡연자의 경우 각각 1.3%, 1.4%로 전 세계적으로 암 관련 사망의 주요 원인으로 5년 생존율이 16%에 불과하다.
* COPD는 흡연자의 최대 50%로 환자의 삶의 질에 현저한 영향을 미친다.
* COPD에서 폐 손상은 산화 스트레스(흡연으로 인한 외인성 및 내인성 모두), 염증성 사이토카인 방출, 프로테아제 활성 및 자가항체 발현에 의해 발생한다.
* COPD는 폐암, 특히 편평 세포 암종에 대한 독립적인 위험 인자로, 정상적인 폐 기능을 가진 흡연자보다 기류 폐쇄가 있는 흡연자에서 발생할 가능성이 최대 5배 더 높다.
* 폐암 발병 연령의 중앙값은 66세이며, COPD는 주로 40세 이상의 흡연자에게 영향을 미치며 60세 이상에서 2.5배 더 높다.

## Chronic obstructive pulmonary disease(COPD) and lung cancer: common pathways for pathogenesis

[논문링크](https://jtd.amegroups.com/article/view/32741/html)

image.png

[그림] COPD 및 폐암 발병으로 이어질 수 있는 유전적 및 후생적 영향<br>
        -- EMT: 상피-간엽 전이, MVB: 다포체, VOC: 휘발성 유기 화합물, EV: 세포외 소포

* COPD와 폐암은 유전학, 후성유전학 및 환경 요인 간의 복잡한 상호 작용으로 이루어진다.
* COPD 환자들 사이에서 폐암 발병률 증가의 정확한 기전은 아직 정의되지 않았다.
* 하지만, 최근 연구에서는 COPD 및 폐암 발병에서 면역 기능 장애, 폐 미생물, 후생 유전학적 조절 및 EV의 역할을 뒷받침하는 증거가 제공되고 있다.
* COPD와 폐암의 주요 분자 경로 사이의 연관성을 조사하는 추가 연구가 필요하다.

## Machine-learning algorithms for asthma, COPD, and lung cancer risk assessment using circulating microbial extracellular vesicle data and their application to assess dietary effects

[논문링크](https://www.nature.com/articles/s12276-022-00846-5)

* 미생물이 난치병에 엄청난 영향을 미친다는 증거가 늘어나고 있지만, 순환하는 미생물 세포외소포(EV)와 호흡기 질환 사이의 관계는 아직 밝혀지지 않았다.
* 호흡기 환자 유래 혈청 미생물 EV metagenomic 데이터의 AI 모델링을 통해 COPD, 천식, 폐암 위험 평가를 위한 새로운 예측 모델 개발과 더불어 고지방식(HFD) 관련 COPD, 천식, 폐암 위험을 줄일 수 있는 식이 보조제를 결정하기 위해 마우스에서 진단 모델을 테스트함.
* 0.93에서 0.99 범위의 평균 AUC 값으로 높은 예측력을 보임
* 마우스 실험에 적용한 결과 다양한 식품이 고지방 식이 관련 천식과 폐암 위험을 감소시키는 반면 COPD는 최소한의 영향을 받는 것으로 나타났다.
* 결론적으로, 천식, COPD 및 폐암 예측에 있어 혈청 미생물 EV는 유용하다.


## Lung Cancer Risk Prediction with Machine Learning Models

[논문링크](https://www.mdpi.com/2504-2289/6/4/139)

* 기계학습 방법을 사용하여 폐암 발병 위험이 높은 개인을 식별하고 장기적인 합병증을 피하기 위한 조기 개입을 위한 효율적인 모델 구축
* Kaggle의 공개데이터셋을 이용. 참가자 수는 309명으로 모든 속성(fatures 15개, target으로 1개)은 다음과 같음:: gender, age, smoking, yellow fingers, anxiety, peer pressure, chronic disease, fatigue, allergy, wheezing, alcohol, coughing, shortness of breath, swallowing difficulty, chest pain, lung cancer
* [kaggle dataset 링크](https://www.kaggle.com/datasets/mysarahmadbhat/lung-cancer/code)
